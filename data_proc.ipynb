```python
import pandas as pd
import csv
import re
```

```python
file_path = "data/realDonaldTrump_in_office.csv"


rows = []
with open(file_path, 'r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file)
    for row in reader:
        combined_column = "".join(row[3:]).replace(",", "")
        new_row = row[:3] + [combined_column]
        rows.append(new_row)

df = pd.DataFrame(rows, columns=["ID", "Time", "Tweet URL", "Tweet Text"])
df = df[['Time','Tweet Text']]
df = df.drop(index=0).reset_index(drop=True)

df.to_csv("data/proc_data.csv", index=True, encoding="utf-8")

print(df.head())
```
